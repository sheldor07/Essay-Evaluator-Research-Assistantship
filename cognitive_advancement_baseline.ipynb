```python
# Define sentences and labels

import pandas as pd

cog_df = pd.read_csv('/content/MIRO week 7&8.csv')
sentences = cog_df["content"]

labels = [
    ["claim"],
    ["elaboration"],
    ["explanation"]
]
```

```python
# Load the zero-shot classification pipeline
classifier = pipeline("zero-shot-classification")
```

```python

# Load the "all-MiniLM-L6-v2" model for sentence embeddings
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')

# Encode sentences to obtain embeddings
sentence_embeddings = model.encode(sentences)


# Define the candidate labels
candidate_labels = ["claim", "elaboration", "explanation"]

# Perform zero-shot classification for each sentence
pred_labels = []
for i, sentence in enumerate(sentences):
    result = classifier(sentence, candidate_labels)
    predicted_label = result["labels"][0]
    predicted_score = result["scores"][0]

    print(f"Sentence {i+1}: {sentence}")
    print(f"Predicted Label: {predicted_label}")
    print(f"Confidence Score: {predicted_score}")
    pred_labels.append(predicted_label)
    print()

cog_df['Predicted Label'] = pred_labels
```

```python
# Save the updated DataFrame to a new CSV file
cog_df.to_csv('predicted_file_cog.csv', index=False)
```

```python
word_to_label = {
    'claim':1,
    'elaboration':2,
    'explaination':3,
}

num_labels = []
word_labels = cog_df['Predicted label']
for label in word_labels:
  num_label = word_to_label[label]
  num_labels.append(num_label)

cog_df['Predicted_Cog_Advancement'] = num_labels
```
