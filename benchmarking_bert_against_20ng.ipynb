```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
```

```python
from sklearn.datasets import fetch_20newsgroups
newsgroups_train = fetch_20newsgroups(subset='train')

from pprint import pprint
pprint(list(newsgroups_train.target_names))
```

```python
categories =['alt.atheism',
 'comp.graphics',
 'comp.os.ms-windows.misc',
 'comp.sys.ibm.pc.hardware',
 'comp.sys.mac.hardware',
 'comp.windows.x',
 'misc.forsale',
 'rec.autos',
 'rec.motorcycles',
 'rec.sport.baseball',
 'rec.sport.hockey',
 'sci.crypt',
 'sci.electronics',
 'sci.med',
 'sci.space',
 'soc.religion.christian',
 'talk.politics.guns',
 'talk.politics.mideast',
 'talk.politics.misc',
 'talk.religion.misc']

newsgroups_train = fetch_20newsgroups(subset='train', categories=categories)

# Initialize an empty list to store the results
results = []

# Perform zero-shot classification on the first 50 documents
for i, document in enumerate(newsgroups_train.data[:5]):
    result = classifier(document, categories)
    results.append((result['labels'][0], result['scores'][0]))  # Store the predicted category and its score
    print(f"Document {i+1} predicted category: {result['labels'][0]} with score: {result['scores'][0]}")
```

```python
# Convert the numerical labels to their corresponding category names
true_labels = [categories[i] for i in newsgroups_train.target[:5]]

# Extract the predicted labels from the results
predicted_labels = [label for label, score in results]

# Compute the accuracy
accuracy = sum(true_label == predicted_label for true_label, predicted_label in zip(true_labels, predicted_labels)) / len(true_labels)

print(f"Accuracy: {accuracy * 100:.2f}%")
```
